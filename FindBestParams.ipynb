```python
import pickle
import numpy as np
import pandas as pd
```

```python
index = []

for a in range(0,10):
    for b in range(0,10):
        index.append('output_files/one_inch_' + str(a) + '_' + str(b) + '.p')
        
df = pd.DataFrame(index=index,columns=['a','b','Max CSO','Tot CSO','Tot WWTP'])

for a in range(0,10):
    for b in range(0,10):
        pickleOut = 'output_files/one_inch_' + str(a) + '_' + str(b) + '.p'
        
        with open(pickleOut,'rb') as f:
            data = pickle.load(f)
            
        CSO = np.array(data['performanceDict']['2908']['ts_flow'])
        WWTP = np.array(data['performanceDict']['1002']['ts_flow'])
        
        df.at['output_files/one_inch_' + str(a) + '_' + str(b) + '.p','a'] = a
        df.at['output_files/one_inch_' + str(a) + '_' + str(b) + '.p','b'] = b
        
        df.at['output_files/one_inch_' + str(a) + '_' + str(b) + '.p','Max CSO'] = max(CSO)
        df.at['output_files/one_inch_' + str(a) + '_' + str(b) + '.p','Tot CSO'] = sum(CSO)
        
        df.at['output_files/one_inch_' + str(a) + '_' + str(b) + '.p','Max WWTP'] = max(WWTP)
        df.at['output_files/one_inch_' + str(a) + '_' + str(b) + '.p','Tot WWTP'] = sum(WWTP)
```

```python
print('WWTP treatment numbers')
print(df['Max WWTP'].min(),df['Max WWTP'].max())
print(df['Tot WWTP'].min(),df['Tot WWTP'].max())

print('CSO numbers')
print(df['Max CSO'].min(),df['Max CSO'].max())
print(df['Tot CSO'].min(),df['Tot CSO'].max())
```
